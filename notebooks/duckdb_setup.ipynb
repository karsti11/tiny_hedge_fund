```python
import duckdb
from datetime import datetime
from src.utils import get_project_root
```

```python
PROJECT_ROOT = get_project_root()
RAW_DATA_FOLDER = PROJECT_ROOT / 'data/raw'
DB_RAW_PATH = RAW_DATA_FOLDER / "raw_layer.db"
DB_PROCESSED_PATH = RAW_DATA_FOLDER / "processed_layer.db"

TODAYS_DATE_STR = datetime.now().date().strftime('%d%m%Y')
COMMODITIES_FILEPATH = RAW_DATA_FOLDER / f'commodities_{TODAYS_DATE_STR}.parquet'
CRYPTO_PRICES_FILEPATH = RAW_DATA_FOLDER / f'crypto_prices_{TODAYS_DATE_STR}.parquet'
```

```python
raw_conn = duckdb.connect(str(DB_RAW_PATH))
```

```python
raw_conn.execute(f"""
    CREATE TABLE commodities AS
    SELECT * FROM read_parquet('{str(COMMODITIES_FILEPATH)}');
""")
```

```python
commodities_df = raw_conn.sql(f"""
    SELECT * FROM commodities;
""").fetchdf()
```
